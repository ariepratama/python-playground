```python
def tokenize(data):
    tokenized = tokenizer(data["text"])
    tokenized["text"] = data["text"]
    return tokenized
```

```python
tokenized_dataset = rotten_tomatoes.map(
    tokenize,
    batched=True,
    remove_columns=["label"]
)
```

```python
chunk_size = 128
```

```python
def group_texts(examples):
    examples["text"] = [ t.split() for t in examples["text"] ]
    # Concatenate all texts
    concatenated_examples = {k: sum(examples[k], []) for k in examples.keys()}
    # Compute length of concatenated texts
    total_length = len(concatenated_examples[list(examples.keys())[0]])
    # We drop the last chunk if it's smaller than chunk_size
    total_length = (total_length // chunk_size) * chunk_size
    # Split by chunks of max_len
    result = {
        k: [t[i : i + chunk_size] for i in range(0, total_length, chunk_size)]
        for k, t in concatenated_examples.items()
    }
    # Create a new labels column
    result["labels"] = result["input_ids"].copy()
    return result
```

```python
lm_dataset = tokenized_dataset.map(group_texts, batched=True)
```

```python
from transformers import DataCollatorForLanguageModeling
```

```python
collator = DataCollatorForLanguageModeling(
    tokenizer,
    mlm=True,
    mlm_probability=0.15,
    return_tensors="pt"
)
```

```python
def collate(batch):
    input_ids = batch["input_ids"]
    collated = collator([input_ids])
    
    return {
        "input_ids": collated["input_ids"],
        "attention_mask": batch["attention_mask"],
        "labels": collated["labels"]
    }
```

```python
import torch.optim as optim
from tqdm.notebook import tqdm
```

```python
collated_dataset = lm_dataset.map(
    collate
)
```

```python
splitted_dataset = collated_dataset.shuffle(
    seed=42
).train_test_split(
    test_size=0.2,
    seed=42
)
```

```python
train, test = splitted_dataset["train"], splitted_dataset["test"]
```

```python
epochs = 10

batch_size = 8
logging_steps = len(train) // batch_size
```

```python
sample = train[0]

sample.keys()
input_ids, attention_mask, labels = [
    torch.tensor(sample[k]).view(1, -1) 
    for k in ["input_ids", "attention_mask", "labels"]
]
labels = labels.long()

lm_model_out = lm_model(input_ids=input_ids, attention_mask=attention_mask, labels=labels.view(-1))

```

```python
optimizer = optim.Adam(lm_model.parameters(), lr=2e-5)
```

```python
for epoch in tqdm(range(epochs)):
    for batch in train:
        input_ids, attention_mask, labels = [
            torch.tensor(batch[k]).view(1, -1) 
            for k in ["input_ids", "attention_mask", "labels"]
        ]
        labels = labels.long()

        lm_model_out = lm_model(
            input_ids=input_ids, 
            attention_mask=attention_mask, 
            labels=labels.view(-1)
        )
        loss = lm_model_out["loss"]
        loss.backward()
        
        optimizer.step()
        
```
